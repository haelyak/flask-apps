```python
#
# reading from a csv file is smooth
#

def get_data_from_csv(csvfilename):
    """ returns a list of all data (list of lists) from a csv file """
    import csv
    all_rows = []
    with open(csvfilename,"r",newline='') as f:
        reader = csv.reader(f)
        for row in reader:
            all_rows.append(row)
    return all_rows

#
# let's try it!
#
if True:
    filename = "tenrows.csv"
    #filename = "wds.csv"
    LoL = get_data_from_csv(filename)            # LoL is a reminder for "list of lists" 
    for item in LoL:
        print(item)
```

```python
#
# change a dictionary of raw counts to a dictionary of percentages (divide by denominator)
#
def get_percentages(d):
    """ should divide each value of a dictionary by the sum of all the values 
        and return a _new_ dictionary with the resulting (fractional) values
        (can multiply by 100 to create percentages: up to you...)
    """
    # this just creates a copy of the original dictionary
    new_d = {}
    sum_of_values = sum(list(d.values()))
    if sum_of_values == 0:
        print(f"get_percentages was provided an input dictionary whose sum-of-values was 0")
        return d       # don't change it

    for k in d:                                 # for each key
        new_d[k] = d[k]/sum_of_values           # assign the original value / sum

    return new_d

#
# let's make sure it's working!
#
if True:
    d = { "y":5, "o":2, "e":3 }   # three letters and their counts...
    new_d = get_percentages(d)
    print(f"    d is {d}\nnew_d is {new_d}")
```

```python

#
# get a list of the items from a dictionary d, sorted:
#

def get_sorted_items(d):
    """ get all of the items from a dictionary, sorted by value
    """
    from operator import itemgetter
    items = d.items()
    sorted_items = sorted(items, key=itemgetter(1), reverse=True)  # sort by value
    # cool! documentation for this is here: https://docs.python.org/3/howto/sorting.html
    return sorted_items

#
# let's try it!
#
if True:
    d = { "y":5, "o":2, "e":3 }   # three letters and their counts...
    SortedItems = get_sorted_items(d)
    print(f"SortedItems: {SortedItems}  ... or ... ")
    for item in SortedItems:
        print(item)
```

```python
#
# make an HTML table from a list-of-lists (by default, saving to the filename "table.html")
#

def make_html_table(LoL, filename):
    """ a very plain html-table creator -- to be improved!  e.g., using
        https://www.w3schools.com/html/html_tables.asp
        https://www.w3schools.com/css/css_table.asp
    """
    header_html = """
     <html>

    <head>
    <style>
    table {
        font-family: Helvetica, sans-serif; /*times, serif; /* /* arial, sans-serif */
        border-collapse: collapse; /* separate */
        width: 50%;
    }

    table tr:hover td {background-color: blueviolet;}

    table tr:nth-child(even) td {background-color: rgb(220, 130, 238);}
    tr:nth-child(odd) td {background-color: rgb(130, 216, 238);}

    td, th {
        border: 2px solid rgb(100,100,100);  /* wow! */
        text-align: left;
        padding: 10px;           /* yikes! */
        background-color: cornflowerblue;
        color: white;
    }

    

    </style>
    </head>

    <body>
    """                           # above is the header HTML

    # heading_html = "<h1> Letters and Their Percentages </h1>"
    table_html = "<table>\n <tr> <th> Letters </th> <th> Percentages </th> </tr>"      # start the table!
    for letter, perc in LoL:
        row_html = f"<tr> <td>{letter}</td> <td>{100*float(perc):7.2f}</td> </tr>\n"   # Wow!
        table_html += row_html
    table_html += "</table>\n"    # end of the table

    footer_html = """
    </body>
    </html>
    """                           # above is the footer HTML

    total_html = header_html + table_html + footer_html  # add up all the strings!

    with open(filename +".html","w") as f:      # open a file for writing, the file's name is f:
        print(total_html, file=f)
        print(f"file {filename} written\n")

    return total_html           # not really needed, since we're ensuring the output goes to file


#
# let's test it
#
if True:
    LoL = get_data_from_csv("tenrows.csv")
    print(f"The table: {LoL}")
    make_html_table( LoL , "table")

```
